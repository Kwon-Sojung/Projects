```python
# 결측치 0으로 변경
df['범죄발생횟수'] = df['범죄발생횟수'].fillna(0)
# 확인
df.isnull().sum()
```

```python
# 데이터 타입 확인
df.dtypes
```

```python
# 범죄발생횟수 수치형데이터로 변경
df["범죄발생횟수"] = pd.to_numeric(df["범죄발생횟수"], errors = "coerce")
# 결측치 0으로 변경
df['범죄발생횟수'] = df['범죄발생횟수'].fillna(0)
# 확인
df.isnull().sum()
df["범죄발생횟수"] = df["범죄발생횟수"].astype(int)

# 데이터 타입확인
df["범죄발생횟수"].dtypes
```

```python
# 소계 하나뿐인지 확인
df.groupby(by=['범죄종류', '범죄명']).count()
```

```python
# '계' 들어가는 행 삭제
drops_col = ['강력범죄', '지능범죄', '폭력범죄', '풍속범죄']
drops = ((df['범죄명'].str.contains('계')) & (df['범죄종류'] == '강력범죄')) |\
((df['범죄명'].str.contains('계')) & (df['범죄종류'] == '폭력범죄')) |\
((df['범죄명'].str.contains('계')) & (df['범죄종류'] == '지능범죄')) |\
((df['범죄명'].str.contains('계')) & (df['범죄종류'] == '풍속범죄')) |\
    df['범죄종류'].str.contains('계') | (df['공범유무']=='계')
df = df[~drops]

#확인
df.sample(10)
```

```python
# 범죄명이 '소계'로 된 컬럼명 바꾸기
drops_col = ['강력범죄', '지능범죄', '폭력범죄', '풍속범죄']
for col in df['범죄종류'].unique().tolist():
    if col not in drops_col:
        df.loc[df['범죄종류'] == col, '범죄명'] = col
```

```python
# 소계 확인
df.groupby(by=['공범유무', '공범관계명']).count()
```

```python
# 공범관계명이 '소계'로 된 컬럼명 바꾸기
drops=["단독범", "미상"]
for col in df["공범유무"].unique().tolist():
    if col in drops:
        df.loc[df["공범유무"]==col, '공범관계명'] = col
```

```python
# 제대로 변경 되었는지 확인
df.groupby(by=['공범유무', '공범관계명']).count()
```

```python
# 공범관계의 공범관계명 중 '계' 들어가는 행 삭제
drops=((df["공범유무"]=="공범관계") & (df["공범관계명"]=="소계"))
df = df[~drops]
#확인
df.groupby(by=["공범유무", "공범관계명"]).count()
```

```python
# 인덱스 재정의
df = df.reset_index(drop=True)
```

```python
# 컬럼 재정렬
cols = ["범죄명", "연도", "공범유무", "공범관계명","범죄종류" , "범죄발생횟수"]
df = df[cols]
```
